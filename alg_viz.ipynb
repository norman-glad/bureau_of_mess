```python
import time
import random
import matplotlib.pyplot as plt

# Simulate the Java methods in Python
def prefixAverage1(x):
    n = len(x)
    a = [0] * n
    for j in range(n):
        total = 0
        for i in range(j + 1):
            total += x[i]
        a[j] = total / (j + 1)
    return a

def prefixAverage2(x):
    n = len(x)
    a = [0] * n
    total = 0
    for j in range(n):
        total += x[j]
        a[j] = total / (j + 1)
    return a

# Generate random input data
def generate_random_array(n):
    return [random.random() for _ in range(n)]

# Measure execution time for a given algorithm and input size
def measure_time(algorithm, x):
    start_time = time.time()
    algorithm(x)
    end_time = time.time()
    return end_time - start_time

# Perform experimental analysis
input_sizes = [10, 100, 1000, 10000]
times1 = []
times2 = []

for n in input_sizes:
    x = generate_random_array(n)
    
    # Measure time for prefixAverage1
    time1 = measure_time(prefixAverage1, x)
    times1.append(time1)
    
    # Measure time for prefixAverage2
    time2 = measure_time(prefixAverage2, x)
    times2.append(time2)
    
    print(f"Input size: {n}, prefixAverage1 time: {time1:.6f} sec, prefixAverage2 time: {time2:.6f} sec")
```

```python
# Plot the results on a log-log chart
plt.figure(figsize=(10, 6))
plt.loglog(input_sizes, times1, marker='o', label='prefixAverage1')
plt.loglog(input_sizes, times2, marker='s', label='prefixAverage2')
plt.xlabel('Input Size (n)')
plt.ylabel('Running Time (seconds)')
plt.title('Running Time Comparison of prefixAverage1 and prefixAverage2')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
```
